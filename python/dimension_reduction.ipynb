```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("poverty_dataset.csv")
df
```

```python
df_17 = df[df['ppp_version'] == 2017]
df_17.reset_index(drop=True, inplace=True)
df_17
```

```python
for i in range(len(df_17.country)):
    if df_17['country'][i] == 'Zimbabwe':
        df_17 = df_17.drop([i])
        df_17 = df_17.drop([i+1])
        break
    if df_17['country'][i+1] == df_17['country'][i]:
        df_17 = df_17.drop([i])
df_17
```

```python
type(df_17)
df_17 = df_17.drop(
    ["reporting_level", "welfare_type", "ppp_version", "survey_year", "survey_comparability"],
    axis=1)
df_17
```

```python
import umap

umap_reducer = umap.UMAP(random_state = 2000)  
# set the seed used by the random number generator to ensure the outcome would be the same each time UMAP is run
umap_df = umap_reducer.fit_transform(df_17.drop(["country", "year"], axis=1))
umap_df
```

```python
import matplotlib.pyplot as plt

scatter = plt.scatter(umap_df[:, 0], umap_df[:, 1])
plt.show()
```

```python
df_17.reset_index(drop=True, inplace=True)
df_17
```

```python
out_df = pd.DataFrame(umap_df)
out_df['country'] = df_17["country"]
out_df['year'] = df_17["year"]
out_df
```

```python
outpath = "umap.csv"
out_df.to_csv(outpath)
```
