```python
import pandas as pd
import sqlite3
```

The following steps are performed with this notebook
1. Read data from file .
2. Make a connection to a sqlite database.
3. Write data into the sqlite database.
4. Close the connection to the database.

### Step 1: Read Data

```python
tree = pd.read_json("./data/tree.json")
map = pd.read_csv("./data/50m.csv")
line = pd.read_csv("./data/line.csv")
umap = pd.read_csv("./data/umap.csv")
mds = pd.read_csv("./data/mds.csv")
tree.head()
```

### Step 2: Connect to database

```python
connection = sqlite3.connect("./data.db")
```

### Step 3: Write to database 
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_sql.html#pandas.DataFrame.to_sql

```python
#tree.to_sql("tree", con=connection, if_exists='replace',
#           index_label='id')
map.to_sql("map", con=connection, if_exists='replace',
           index_label='id')
line.to_sql("line", con=connection, if_exists='replace',
           index_label='id')
umap.to_sql("umap", con=connection, if_exists='replace',
           index_label='id')
mds.to_sql("mds", con=connection, if_exists='replace',
           index_label='id')
cur = connection.execute("SELECT * from umap")
for row in cur:
    print(row)
```

### Step 4: Close database connection

```python
connection.close()
```
